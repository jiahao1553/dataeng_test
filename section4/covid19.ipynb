```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# retrieve data
confirmed_json = requests.get("https://api.covid19api.com/country/singapore/status/confirmed").json()
death_json = requests.get("https://api.covid19api.com/country/singapore/status/deaths").json()
recovered_json = requests.get("https://api.covid19api.com/country/singapore/status/recovered").json()
```

```python
# convert json to dataframe and parse date
confirmed = pd.DataFrame.from_dict(confirmed_json)
deaths = pd.DataFrame.from_dict(death_json)
recovered = pd.DataFrame.from_dict(recovered_json)

confirmed['Date'] = pd.to_datetime(confirmed['Date'], format = '%Y-%m-%dT%H:%M:%SZ')
deaths['Date'] = pd.to_datetime(deaths['Date'], format = '%Y-%m-%dT%H:%M:%SZ')
recovered['Date'] = pd.to_datetime(recovered['Date'], format = '%Y-%m-%dT%H:%M:%SZ')
```

```python
# plot line chart
plt.plot(confirmed['Date'], confirmed['Cases'], color='blue')
plt.plot(deaths['Date'], deaths['Cases'], color='red')
plt.plot(recovered['Date'], recovered['Cases'], color='green')
plt.legend(["Confirmed","Deaths","Recovered"])
plt.title('No. of Covid-19 Cases in Singapore', fontsize=12)
plt.xlabel('Date', fontsize=12)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=90)
plt.ylabel('No. of Cases', fontsize=14)
plt.show()
```
